## 2.1 Integrales y ecuaciones diferenciales

La evaluación numérica de integrales de la forma

$\displaystyle \int_a^b f(x) dx$

se llaman *cuadratura numérica*, o simplemente *cuadratura*. SciPy tiene las funciones  `quad`, `dblquad` y `tplquad` para cuadraturas simples, dobles o triples.

```python
from scipy.integrate import quad

quad?
```

```python
# definimos f(x)
def f(x):
    return 2*x


val, abserr = quad(f, 0, 3)
print("integral value =", val, ", absolute error =", abserr)
```

#### Integración de funciones con parámetros adicionales


Si necesitaramos pasarle parámetros a nuestra función lo hacemos en un tupla en el argumento `args`. Por ejemplo, supongamos que tenemos una función de la forma $g(x) = x^n$ donde `x` es la variable independiente y `n` es un parámetro.

```python
def g(x, n):
    return x**n
```

Si quisieramos evaluar la integral de esta función de x fijando n en 3

```python
# evalua la integral x**3
quad(g, 0, 5, args=(3,))
```

#### Ecuaciones diferenciales ordinarias

El paquete de integración también tiene funciones para resolver [ecuaciones diferenciales ordinarias](https://en.wikipedia.org/wiki/Ordinary_differential_equation)

Por ejemplo, para resolver 

$$\frac{\mathrm{d}y}{\mathrm{d}t} = -2y$$

 con $t  \in [-1, 5]$  y la condición inicial $y(0) = 1$ 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
t = np.linspace(-1, 5, 100)

y = odeint(lambda y,t: -2*y, 1, t)    # lamda is a function that is defined without a name

plt.plot(t,y)
plt.show()
```

```python
def f(y, t): 
    return -2 * y

t = np.linspace(-1, 5, 100)

y = odeint(f, 1, t)

plt.plot(t,y)
plt.show()
```

## 2.2 Álgebra lineal

El modulo de álgebra lineal de scipy es bastante más eficiente que el de numpy. Entre otras operaciones comunes podemos calcular el determinante de una matriz, la inversa, aplicar una descomposición de valor singular (SVD).

```python
import numpy as np
from scipy import linalg

arr = np.array([[1, 2],[3, 4]])
print(linalg.det(arr))

arr = np.array([[3, 2],[6, 4]])
print(linalg.det(arr))
```

```python
linalg.det(np.ones((3, 4))) # Ups !
```

Invertir

```python
arr = np.array([[1, 2], [3, 4]])
iarr = linalg.inv(arr)
print(iarr)
print(np.allclose(np.dot(arr, iarr), np.eye(2))) #funcion allclose verifica que dos arrays sean iguales
#np.eye matriz identidad 
```

Si queremos calcular la inversa de una matriz singular (es decir que su determinante es cero) va a gritar una excepción LinAlgError:

```python
arr = np.array([[3, 2],[6,4]])
linalg.inv(arr) # Ups !
```

Calculamos la descomposición SVD (Singular value decomposition)

```python
arr = np.arange(9).reshape((3, 3)) + np.diag([1, 0, 1])

# m, n = 9, 6
# arr = np.random.randn(m, n) + 1.j*np.random.randn(m, n)
```

```python
uarr, spec, vharr = linalg.svd(arr)

# uarr: matriz que tiene los autovectores (izq) como columnas
# spec: vector con los autovectores
# vharr: matriz que tiene los autovectores (der) como filas

print(arr.shape)
print(spec.shape)
print(vharr.shape)


sarr = np.diag(spec)
svd_mat = np.dot(uarr, np.dot(sarr, vharr))

np.allclose(svd_mat, arr)
```

La descoposición SVD se utilzia comunmente en estadísitca y procesado de señales. Sin embargo, existen muchas otras descomposiciones en *scipy* como (QR, LU, Cholesky, Schur), como distintas formas de resolver sistemas lineales.

## 2.3 Optimización y ajuste de curvas

### Encontrar mínimos

Supongamos que tenemos una función cualquiera de una o más variables

```python
def fpol(x):
    return x**4 + 5*x**3 + (x-2)**2
```

```python
fig, ax  = plt.subplots()
x = np.linspace(-5, 2, 100)
ax.plot(x, fpol(x))
plt.show()
```

Podemos encontrar un mínimo local con la función minimize, que recibe como parámetro la función objetivo y un valor de estimación inicial a partir del cual se comienza a "ajustar"

```python
from scipy import optimize
result = optimize.minimize(fpol, -3, method='BFGS') # http://en.wikipedia.org/wiki/BFGS
result
```

```python
result['x']
```

```python
fig, ax  = plt.subplots()
x = np.linspace(-5, 2, 100)
ax.plot(x, fpol(x))
ax.plot(result['x'], result['fun'], 'or')
plt.show()
```

`minimize` es una "función paraguas" para diferentes algoritmos, que se pueden elegir mediante el parámetro `method` o bien utilizar la función explícita de minimización que lo implementa.


```python
result = optimize.minimize(fpol, 0, method='BFGS')
result['x']
```

```python
fig, ax  = plt.subplots()
x = np.linspace(-5, 2, 100)
ax.plot(x, fpol(x))
ax.plot(result['x'], result['fun'], 'or')
plt.show()
```

Ojo que hay métodos que son muy sensibles al valor de estimación

```python
optimize.minimize(fpol, 0, method='Powell')['x']
```

Vemos que tiene forma de exponencial decreciente, es decir

$$
    f(x) = ae^{-bx} + c
$$

El objetivo es encontrar cuales son los valores de los parámetros `a`, `b` y `c` que mejor ajustan la curva a los puntos dados. 


```python
from scipy import optimize

def func(x, a, b, c):   # x es la variable independiente, a b y c los parametros a encontrar
    return a*np.exp(-b*x) + c

popt, pcov = optimize.curve_fit(func, x, y)
print(popt)
# perr = np.sqrt(np.diag(pcov))
# print(perr)
```

Por lo tanto la curva que mejor ajusta a nuestros datos es

$$f(x) = 2.33e^{-1.18x} + 0.45$$

```python
fitted = func(x, *popt)

plt.figure(figsize=(9,6))
plt.scatter(x, y, marker='.')
plt.plot(x, fitted, 'r', label='fitted')
plt.legend()

plt.show()
```

## 2.5 Interpolación


El módulo de interpolación ( `scipy.interpolate`) es muy útil para aproximar una función de los datos experimentales . Este módulo se basa en la librería FITPACK del proyecto netlib implementada en Fortran.


Veamos un ejemplo, en este caso vamos a ajustar una funcion seno a nuestros datos

```python
measured_time = np.linspace(0, 1, 10)
noise = (np.random.random(10)*2 - 1) * 1e-1
measures = np.sin(2 * np.pi * measured_time) + noise
```

```python
plt.plot(measured_time, measures, 'o', ms=6, label='measures')
plt.legend()
plt.show()
```

Calculemos una interpolación lineal de los datos mediante la clase `scipy.interpolate.interp1d`

```python
from scipy.interpolate import interp1d
linear_interp = interp1d(measured_time, measures)
```

Ahora evaluemos la función en el rango [0,1]

```python
computed_time = np.linspace(0, 1, 50)
linear_results = linear_interp(computed_time)
```

```python
plt.plot(measured_time, measures, 'o', ms=6, label='measures')
plt.plot(computed_time, linear_results, label='linear interp')
plt.legend()
plt.show()
```

Pero también podemos realizar una interpolación cúbica indicando esto en el argumento `kind` de la siguiente forma

```python
cubic_interp = interp1d(measured_time, measures, kind='cubic')
cubic_results = cubic_interp(computed_time)
```

Ahora veamos que interpolación es mejor para aproximar los datos que tenemos

```python
plt.plot(measured_time, measures, 'o', ms=6, label='measures')
# plt.plot(computed_time, linear_results, label='linear interp')
plt.plot(computed_time, cubic_results, label='cubic interp')
plt.legend()
plt.show()
```

## 2.6 Un poco de procesamiento de imágenes

Una imágen digital tipo [raster](https://en.wikipedia.org/wiki/Raster_graphics) (matriz de pixeles) puede ser facilmente interpretable como un array multidimensional. Por ejemplo, una imágen de 512x512 pixeles a color puede alamcenarse en una "matriz" de 512x512x3 donde cada una las 3 matrices representa una capa de colo RGB (rojo, verde, azul). Para una imágen monocromática, obviamente una sóla matriz nos alcanza. 

El módulo dedicado al procesado de imágenes en *scipy* es `scipy.ndimage`.

```python
from scipy import ndimage
```

```python
from scipy.misc import face
image =  face()
print(image.shape)
print(type(image))
```

```python
plt.imshow(image)
plt.show()
```

```python
image_gray = image[:,:,1].astype(np.float64)

plt.imshow(image_gray, cmap = 'gray')
plt.show()
```

Ahora bien, el módulo de manipulación de **imágenes multidimensionales** `scipy.ndimage`  tiene un montón de funciones útiles como pueden ser la convolución, FFT, distintos filtros, operaciones morfológicas y transformaciónes geométricas. Veamos algunas

```python
plt.imshow(ndimage.convolve(image_gray, np.ones((21,21))), cmap='gray')
plt.show()
```

```python
shifted_img = ndimage.shift(image_gray, (50, 50))
shifted_img2 = ndimage.shift(image_gray, (50, 50), mode='nearest')

rotated_img = ndimage.rotate(image_gray, 30)

cropped_img = image_gray[150:-200, 250:-100]

zoomed_img = ndimage.zoom(image_gray, 2)

imgs = (shifted_img, shifted_img2, rotated_img, cropped_img, zoomed_img)

fig, axes = plt.subplots(1, 5, figsize=(20,100))

for ax, img in zip(axes, imgs):
    ax.tick_params(left=False, bottom=False, labelleft=False, labelbottom=False)
    ax.imshow(img, cmap='gray')
plt.show()
```
